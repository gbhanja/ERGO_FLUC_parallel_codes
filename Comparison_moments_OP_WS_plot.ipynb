#### Extracting work statistics data

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import os

downloads = os.path.join(os.path.expanduser("~"), "Downloads")

filename = os.path.join(downloads, "Fluc_erg_WS_all_moments.npz")

data_WS = np.load(filename)

# extract arrays
N_arr = data["N"]
tau = data["tau"]
Eb = data["Eb"]
Eerg = data["Eerg"]
ratio = data["ratio"]
variance = data["variance"]
third_moment = data["third_moment"]
fourth_moment = data["fourth_moment"]

# derived quantities
F_ratio = np.sqrt(variance) / Eerg
FE_ratio = np.sqrt(variance) / Eb
```

```python
for i in range(len(N_arr)):
    print(
        f"N={N_arr[i]}, "
        f"τ={tau[i]:.4f}, "
        f"Eb={Eb[i]:.6f}, "
        f"Ergotropy={Eerg[i]:.8f}, "
        f"Ratio={ratio[i]:.8f}, "
        f"Variance={variance[i]:.10e},"
        f"Third Moment={third_moment[i]:.12e},"
        f"Fourth Moment={fourth_moment[i]:.12e},"
        f"Fluctuation w.r.t erg={F_ratio[i]:.12e},"
        f"Fluctuation w.r.t EB={FE_ratio[i]:.12e}"
    )
```

```python
E_ergo_OP = data_OP["Eerg"]
E_var_OP = data_OP["variance"]
E_third_OP = data_OP["third_moment"]
E_fourth_OP = data_OP["fourth_moment"]

E_ergo_WS = data_WS["Eerg"]
E_var_WS = data_WS["variance"]
E_third_WS = data_WS["third_moment"]
E_fourth_WS = data_WS["fourth_moment"]

erg_OP = np.array(E_ergo_OP).flatten()
var_OP = np.array(E_var_OP).flatten()
third_OP = np.array(E_third_OP).flatten()
fourth_OP = np.array(E_fourth_OP).flatten()

erg_WS = np.array(E_ergo_WS).flatten()
var_WS = np.array(E_var_WS).flatten()
third_WS = np.array(E_third_WS).flatten()
fourth_WS = np.array(E_fourth_WS).flatten()
```

```python
######################################
# Plot 1: Ergotropy (1st Moment) vs N
######################################
plt.figure(figsize=(8, 5))
plt.plot(N_arr, erg_OP, 'o-', color='blue', label='OP Formalism', markersize=6)
plt.plot(N_arr, erg_WS, 's--', color='red', label='WS Formalism', markersize=3)

plt.xlabel(r"$N$ (Number of spins)", fontsize=12)
plt.ylabel(r"Ergotropy ($\mathcal{E})$", fontsize=12)
plt.title(r"Ergotropy vs N", fontsize=14)
plt.legend()
plt.tight_layout()
#plt.savefig("m1.png")
plt.show()
```

```python
#######################################
# Plot 2: Variance (2nd Cumulant) vs N
#######################################
plt.figure(figsize=(8, 5))
plt.plot(N_arr, var_OP, 'o-', color='blue', label='OP Formalism', markersize=6)
plt.plot(N_arr, var_WS, 's--', color='red', label='WS Formalism', markersize=3)

plt.xlabel(r"$N$ (Number of spins)", fontsize=12)
plt.ylabel(r"$\mathrm{Var}(\mathcal{E})$", fontsize=12)
plt.title(r"Variance vs N", fontsize=14)
plt.legend()
plt.tight_layout()
#plt.savefig("m2.png")
plt.show()
```

```python
############################
# Plot 3: Third Moment vs N
############################
plt.figure(figsize=(8, 5))
plt.plot(N_arr, third_OP, 'o-', color='blue', label='OP Formalism', markersize=6)
plt.plot(N_arr, third_WS, 's--', color='red', label='WS Formalism', markersize=3)

plt.xlabel(r"$N$ (Number of spins)", fontsize=12)
plt.ylabel(r"Third Moment of $\mathcal{E}$", fontsize=12)
plt.title(r"Third Moment of $\mathcal{E}$ vs N", fontsize=14)
plt.legend()
plt.tight_layout()
#plt.savefig("m3.png")
plt.show()
```

```python
#############################
# Plot 4: Fourth Moment vs N
#############################
plt.figure(figsize=(8, 5))
plt.plot(N_arr, fourth_OP, 'o-', color='blue', label='OP Formalism', markersize=6)
plt.plot(N_arr, fourth_WS, 's--', color='red', label='WS Formalism', markersize=3)

plt.xlabel(r"$N$ (Number of spins)", fontsize=12)
plt.ylabel(r"Fourth Moment of $\mathcal{E}$", fontsize=12)
plt.title(r"Fourth Moment of $\mathcal{E}$ vs N", fontsize=14)
plt.legend()
plt.tight_layout()
#plt.savefig("m4.png")
plt.show()
```

```python
# Absolute differences
diff_third = third_WS - third_OP
diff_fourth = fourth_WS - fourth_OP

##############################################
# Plot 5: Absolute Difference in Third Moment
##############################################

plt.figure(figsize=(8, 5))
plt.plot(N_arr, diff_third, 'p-', color='purple', markersize=5)
plt.xlabel(r"$N$ (Number of spins)", fontsize=11)
plt.ylabel(r"$\Delta \mu_3$ (WS $-$ OP)", fontsize=11)
plt.title("Absolute Separation in Third Moment", fontsize=12)
plt.tight_layout()
plt.show()


##############################################
# Plot 6: Absolute Difference in Fourth Moment 
##############################################

plt.figure(figsize=(8, 5))
plt.plot(N_arr, diff_fourth, 'd-', color='forestgreen', markersize=5)
plt.xlabel(r"$N$ (Number of spins)", fontsize=11)
plt.ylabel(r"$\Delta \mu_4$ (WS $-$ OP)", fontsize=11)
plt.title("Absolute Separation in Fourth Moment", fontsize=12)
plt.tight_layout()
plt.show()
```
